```python
# mengimport semua library yang dibutuhkan

import cv2 as cv
import numpy as np
from PIL import Image
import math
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from math import log10, sqrt
import glob
```

1. melakukan Gamma Correction

```python
print(' Gamma Correction pada citra ')
print('----------------------------------')
try:
  gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
  print('Error, not a number')

image = cv.imread('/content/drive/MyDrive/content/female.tiff')
gamma_image = np.zeros(image.shape, image.dtype)

for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    for c in range(image.shape[2]):
      gamma_image[y,x,c] = np.clip(255 * pow(image[y,x,c]/255, (1/gamma)), 0, 255)

final_frame = cv.hconcat((image, gamma_image))
cv2_imshow(final_frame)
```

2. Melakukan simulasi image depth

```python
bit_depth = 2
level = 255 / (pow(2,bit_depth)-1)
image = cv.imread('/content/drive/MyDrive/content/female.tiff', cv.IMREAD_GRAYSCALE)
image_depth = np.zeros(image.shape, image.dtype)

for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    image_depth[y,x] = int(image[y, x] / level) * int(level)

final_frame = cv.hconcat((image, image_depth))
cv2_imshow(final_frame)
```

3. Melakukan praktikum avarage denoising

a. Jumlah citra avarage 5

```python
cv_img = []
for img in glob.glob('/content/drive/MyDrive/content/noises/*.jpg'):
  n = cv.imread(img)
  cv_img.append(n)

ori = cv.imread('/content/drive/MyDrive/content/galaxy.jpg')
noise_image = cv_img[4]

for i in range(4):
  alpha = 1.0 / (i + 1)
  beta = 1.0 - alpha
  noise_image = cv.addWeighted(cv_img[i], alpha, noise_image, beta, 0.0)

def PSNR(ori, compressed):
    mse = np.mean((ori - compressed) ** 2)
    if(mse == 0):  # MSE is zero means no noise is present in the signal .
                  # Therefore PSNR have no importance.
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr

compressed = noise_image
psnr = PSNR(ori, compressed)
print(psnr , ' db')

final_frame = cv.hconcat((ori, noise_image))
cv2_imshow(final_frame)
```

4. Image Masking

```python
img = cv.imread('/content/drive/MyDrive/content/couple.tiff')
mask_ori = np.zeros(img.shape, dtype=np.uint8)
mask_ori = cv.circle(mask_ori, (70, 50), 50, (255,255,255), -1)

mask2 = np.zeros(img.shape, dtype=np.uint8)
mask2 = cv.circle(mask_ori, (190, 50), 50, (255,255,255), -1)

result = img & mask_ori & mask2
result[mask_ori==0] = 255

final_frame = cv.hconcat((img, mask_ori, result))
cv2_imshow(final_frame)
```
